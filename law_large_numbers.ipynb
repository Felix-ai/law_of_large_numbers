```python
# Import Modules
import numpy as np
import random
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('ggplot')
```

```python
# Parameters
N = 1000
seq = range(1, 7)
```

```python
# Function
def sample_mean(s, k):
    return np.mean(random.choices(s, k=k))
```

```python
# Sample means
x_bars = []
for n in range(1, N):
    x_bars.append(sample_mean(seq, n))
```

```python
# Plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.axhline(np.mean(seq), color="k")
ax.plot(range(1, N), x_bars, alpha=0.8)
ax.set_xlabel('Number of Trials', color="b")
ax.set_xlim(0, N)
ax.set_ylabel('Average', color="b")
ax.set_ylim(min(seq), max(seq))
ax.set_title('Average dice roll by number of rolls')
ax.legend(["Theoretical mean", "Obserevd averages"])
plt.savefig("images/law_large_numbers.png", bbox_inches="tight")
```
